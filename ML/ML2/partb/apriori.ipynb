```python
import numpy as np
import pandas as pd
data=[['pen','pencil','eraser','glitter'],
      ['pencil','eraser','glitter','highlighter'],
      ['pen','pencil','eraser','highlighter'],
      ['pen','eraser','glitter','highlighter'],
      ['pencil','eraser','highlighter'],
      ['pen','pencil','eraser','glitter']]
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_ary=te.fit(data).transform(data)
df=pd.DataFrame(te_ary, columns=te.columns_)
df.head(5)
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
freq_itemsets = apriori(df, min_support=0.10, use_colnames=True)
freq_itemsets
```

```python
rls=association_rules(freq_itemsets, metric='confidence', min_threshold=1)
rls
```

```python
rls.head(3)
```

```python
rls1=rls[['antecedents','consequents','support','confidence']]
rls2=rls1[rls1['confidence']>=0.9]
rls2
```
