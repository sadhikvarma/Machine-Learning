```python
import cv2

def extract_edges(image_path):
    # Read the image from the given path
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    # Apply Canny edge detection
    edges = cv2.Canny(image, 100, 200)
    return edges

input_image_path = 'image.jpg'
# Extract edges from the image
edges_image = extract_edges(input_image_path)
# Save the edges image
cv2.imwrite('edges_output.jpg', edges_image)
# Display the original and edges images (optional)
cv2.imshow('Original Image',cv2.imread(input_image_path))
cv2.imshow('Edges Image',edges_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
def sobel_edge_detection(image):
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # Apply Sobel operators to get gradients in the horizontal and vertical directions
    sobel_x = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=3)
    sobel_y = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=3)
    # Calculate the gradient magnitude
    gradient_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)

    # Normalize the gradient magnitude to 0-255 range
    gradient_magnitude = np.uint8(255 * gradient_magnitude / np.max(gradient_magnitude))

    # Apply a threshold to obtain binary edges
    threshold_value = 100
    _, edges = cv2.threshold(gradient_magnitude, threshold_value, 255, cv2.THRESH_BINARY)
    return edges
# Read the input image
input_image = cv2.imread('image.jpg')

# Apply Sobel edge detection
edges = sobel_edge_detection(input_image)
# Display the original and edge-detected images
cv2.imshow('Original Image', input_image)
cv2.imshow('Sobel Edge Detection', edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

def laplacian_edge_detection(image):
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply Laplacian edge detection
    laplacian = cv2.Laplacian(gray_image, cv2.CV_64F)

    # Convert the result to unsigned 8-bit integer
    laplacian = cv2.convertScaleAbs(laplacian)

    return laplacian

# Read the input image
input_image = cv2.imread('sample.jpg')

# Apply Laplacian edge detection
edges = laplacian_edge_detection(input_image)

# Display the original and edge-detected images
cv2.imshow('Original Image', input_image)
cv2.imshow('Laplacian Edge Detection', edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
 
```

```python
import cv2
import numpy as np

def prewitt_edge_detection(image):
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply Prewitt operators to get gradients in the horizontal and vertical directions
    kernel_x = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
    kernel_y = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
    prewitt_x = cv2.filter2D(gray_image, -1, kernel_x)
    prewitt_y = cv2.filter2D(gray_image, -1, kernel_y)

    # Calculate the gradient magnitude
    gradient_magnitude = np.sqrt(prewitt_x**2 + prewitt_y**2)

    # Normalize the gradient magnitude to 0-255 range
    gradient_magnitude = np.uint8(255 * gradient_magnitude / np.max(gradient_magnitude))
    # Apply a threshold to obtain binary edges
    threshold_value = 100
    _, edges = cv2.threshold(gradient_magnitude, threshold_value, 255, cv2.THRESH_BINARY)

    return edges
# Read the input image
input_image = cv2.imread('image.jpg')

# Apply Prewitt edge detection
edges = prewitt_edge_detection(input_image)

# Display the original and edge-detected images
cv2.imshow('Original Image', input_image)
cv2.imshow('Prewitt Edge Detection', edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

def roberts_cross_edge_detection(image):
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply Roberts Cross operators to get gradients in the horizontal and vertical directions
    kernel_x = np.array([[1, 0], [0, -1]])
    kernel_y = np.array([[0, 1], [-1, 0]])
    roberts_x = cv2.filter2D(gray_image, -1, kernel_x)
    roberts_y = cv2.filter2D(gray_image, -1, kernel_y)

    # Calculate the gradient magnitude
    gradient_magnitude = np.sqrt(roberts_x**2 + roberts_y**2)

    # Normalize the gradient magnitude to 0-255 range
    gradient_magnitude = np.uint8(255 * gradient_magnitude / np.max(gradient_magnitude))

    # Apply a threshold to obtain binary edges
    threshold_value = 100
    _, edges = cv2.threshold(gradient_magnitude, threshold_value, 255, cv2.THRESH_BINARY)

    return edges

# Read the input image
input_image = cv2.imread('sample.jpg')

# Apply Roberts Cross edge detection
edges = roberts_cross_edge_detection(input_image)

# Display the original and edge-detected images
cv2.imshow('Original Image', input_image)
cv2.imshow('Roberts Cross Edge Detection', edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
