```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import matplotlib.pyplot as plt

# Load the Iris dataset for demonstration
iris = load_iris()
X = iris.data
y = iris.target

# Perform LDA
lda = LinearDiscriminantAnalysis(n_components=2)  # You can specify the number of components you want
X_lda = lda.fit_transform(X, y)

# Plot LDA-transformed data
plt.scatter(X_lda[:, 0], X_lda[:, 1], c=y, cmap='viridis')
plt.xlabel('LDA Component 1')
plt.ylabel('LDA Component 2')
plt.title('LDA-Transformed Data')
plt.colorbar(label='Class')

plt.show()
```

```python
import pandas as pd
from sklearn.decomposition import TruncatedSVD, FactorAnalysis
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris


# Load the dataset
data = load_iris()
# convert to csv
X, y = data.data, data.target

svd = TruncatedSVD(n_components=2)
X_svd = svd.fit_transform(X)

# Factor Analysis (FA)
fa = FactorAnalysis(n_components=2)
X_fa = fa.fit_transform(X)

# Visualize the results
plt.figure(figsize=(12, 5))



# SVD
plt.subplot(1, 2, 1)
plt.scatter(X_svd[:, 0], X_svd[:, 1], c=data['target'], cmap='viridis')
plt.title(f'SVD (n_components={2})')
plt.xlabel('Component 1')
plt.ylabel('Component 2')

# FA
plt.subplot(1, 2, 2)
plt.scatter(X_fa[:, 0], X_fa[:, 1], c=data['target'], cmap='viridis')
plt.title(f'Factor Analysis (n_components={2})')
plt.xlabel('Factor 1')
plt.ylabel('Factor 2')

plt.tight_layout()
plt.show()
```
