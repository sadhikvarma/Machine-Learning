```python
from mahotas.features import surf
import mahotas as mh
import matplotlib.pyplot as plt
from skimage import io
cat=io.imread('cat.webp')
dog=io.imread('dog.jpg')
cat_mh = mh.colors.rgb2gray(cat)
dog_mh = mh.colors.rgb2gray(dog)
cat_surf = surf.surf(cat_mh, nr_octaves=8, nr_scales=16, initial_step_size=1,threshold=0.1, max_points=40)
dog_surf = surf.surf(dog_mh, nr_octaves=8, nr_scales=16, initial_step_size=1,threshold=0.1, max_points=40)
fig = plt.figure(figsize = (10,4))
ax1 = fig.add_subplot(1,2, 1)
ax1.imshow(surf.show_surf(cat_mh, cat_surf))
ax2 = fig.add_subplot(1,2, 2)
ax2.imshow(surf.show_surf(dog_mh, dog_surf))
```

```python
import cv2
import mahotas as mh
def extract_sift_features(image_path):
    # Load the image in grayscale
    img_gray = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Initialize the SIFT detector
    sift = cv2.SIFT_create()

    # Detect key points and compute SIFT descriptors
    keypoints, descriptors = sift.detectAndCompute(img_gray, None)

    # Draw the keypoints on the image
    img_with_keypoints = cv2.drawKeypoints(img_gray, keypoints, None)

    # Display the image with keypoints
    cv2.imshow('Image',img_with_keypoints)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    # Return the keypoints and descriptors
    return keypoints, descriptors

# Replace 'sample_image.jpg' with the path to your sample image
sample_image_path = 'dog.jpg'
keypoints, descriptors = extract_sift_features(sample_image_path)
```
