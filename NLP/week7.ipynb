```python
import nltk
from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt

# Sample text for testing
text = "Stemming algorithms aim to remove suffixes (and in some cases prefixes) from a word in order to obtain a root form of the word."

# Tokenize the text
tokens = word_tokenize(text)

# Initialize stemmers
porter_stemmer = PorterStemmer()
lancaster_stemmer = LancasterStemmer()
snowball_stemmer = SnowballStemmer("english")

# Perform stemming using different techniques
stemming_methods = ['Porter Stemmer', 'Lancaster Stemmer', 'Snowball Stemmer']
stemmed_texts_porter = [porter_stemmer.stem(token) for token in tokens]
stemmed_texts_lancaster = [lancaster_stemmer.stem(token) for token in tokens]
stemmed_texts_snowball = [snowball_stemmer.stem(token) for token in tokens]

# Prepare data for chart
unique_stems_porter = len(set(stemmed_texts_porter))
unique_stems_lancaster = len(set(stemmed_texts_lancaster))
unique_stems_snowball = len(set(stemmed_texts_snowball))
unique_stems = [unique_stems_porter, unique_stems_lancaster, unique_stems_snowball]

# Plotting
plt.bar(stemming_methods, unique_stems, color=['blue', 'green', 'red'])
plt.xlabel('Stemming Technique')
plt.ylabel('Unique Stems Count')
plt.title('Performance of Stemming Techniques')
plt.show()
```
