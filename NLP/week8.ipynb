```python
import nltk
import spacy
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt

# Sample text for testing
text = "Lemmatisation (or lemmatization) is the process of grouping together the inflected forms of a word so they can be analysed as a single item."

# Tokenize the text
tokens = word_tokenize(text)

# Initialize lemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
nlp = spacy.load('en_core_web_sm')

# Perform lemmatization using different techniques
lemmatization_methods = ['WordNet Lemmatizer', 'Spacy Lemmatizer']
lemmatized_texts_wordnet = [wordnet_lemmatizer.lemmatize(token) for token in tokens]
lemmatized_texts_spacy = [token.lemma_ for token in nlp(text)]

# Prepare data for chart
unique_lemmas_wordnet = len(set(lemmatized_texts_wordnet))
unique_lemmas_spacy = len(set(lemmatized_texts_spacy))
unique_lemmas = [unique_lemmas_wordnet, unique_lemmas_spacy]

# Plotting
plt.bar(lemmatization_methods, unique_lemmas, color=['blue', 'green'])
plt.xlabel('Lemmatization Technique')
plt.ylabel('Unique Lemmas Count')
plt.title('Performance of Lemmatization Techniques')
plt.show()
```
